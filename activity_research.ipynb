**Исследование датасета за 4.10.2024**

```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt

# 1. Монтируем Google Drive и загружаем данные
drive.mount('/content/drive')

# Укажите путь к вашему файлу
file_path = '/content/drive/MyDrive/StrangerThings/dataset/data_2024-10-04.parquet'

# Читаем файл
try:
    f4 = pd.read_parquet(file_path)
    print(f"Файл {file_path} успешно загружен. Строк: {len(f4)}")

    # Конвертируем ts в datetime, если это еще не сделано
    f4['ts'] = pd.to_datetime(f4['ts'])
    f4.info()
except Exception as e:
    print(f"Ошибка при чтении файла: {e}")
    raise

```

```python
import seaborn as sns

# Преобразование столбца 'ts' в datetime и извлечение часа
f4['hour'] = pd.to_datetime(f4['ts']).dt.hour  # Извлекаем час (0-23)

# Подсчет событий по часам
hourly_activity = f4['hour'].value_counts().sort_index()

# Визуализация
plt.figure(figsize=(14, 6))
ax = sns.barplot(x=hourly_activity.index, y=hourly_activity.values, color='royalblue')

# Настройка графика
plt.title('Активность пользователей по часам (4.10.2024)', fontsize=16, pad=20)
plt.xlabel('Час дня', fontsize=12)
plt.ylabel('Количество событий', fontsize=12)

# Форматирование оси X (все часы от 0 до 23)
plt.xticks(range(0, 24), [f'{h}:00' for h in range(24)], rotation=45)

# Добавляем значения на столбцы
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 5),
                textcoords='offset points')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Отображение графика
plt.show()
```
